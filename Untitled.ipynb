```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np



#study data files
Overdose_by_opiods_df = pd.read_csv('raw_data.csv')
age_usage_df = pd.read_csv('drug-use-by-age.csv')
legal_by_year_df = pd.read_csv('Marijuana legality by state.xlsx.csv')
overdoses_by_year_df = pd.read_csv('VSRR_Provisional_Drug_Overdose_Death_Counts.csv')
#legal_by_year = pd.read_csv('legalbyyear.csv')


Overdose_by_opiods_df =Overdose_by_opiods_df.rename(columns={'Natural and Semisynthetic Opioids (e.g. oxycodone, hydrocodone)': 'Natural_Opioid', 'Synthetic Opioids, other than Methadone (e.g. fentanyl, tramadol)': 'Synthetic_Opioid'})









```
